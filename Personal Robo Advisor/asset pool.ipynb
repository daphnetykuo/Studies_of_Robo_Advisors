```python
stock_etf = ['VTI', 'VGK', 'VPL', 'VWO', 'PFF']
bond_etf = ['IEI', 'BWX','AGG','BND']
curr_etf = ['FXY','FXF','UDN']
comm_etf = ['IAU','AOA','AOK','USO']

 ['VTI', 'VGK', 'VPL', 'VWO', 'PFF','IEI', 'BWX','AGG','BND','FXY','FXF','UDN','IAU','AOA','AOK','USO']
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import datetime as dt
```

```python
commodity = pd.DataFrame()
comm_id = comm_etf
q = len(comm_id)
records = 0
for i in comm_id:
    tmp = web.DataReader(i, 'yahoo', '1/2/2014', '4/30/2019')   #1/3 is the first business day of 2005
    commodity[i] = tmp['Adj Close']


commodity.head()
```

```python
comm_ret = np.log(commodity / commodity.shift(1)).dropna()
comm_ret.head()
```

```python
plt.figure(figsize=(12,8))
for i in comm_id:
    plt.plot(commodity[i], label=i)
    
plt.legend(loc='upper left', fontsize=12)
```

```python
tickers = ['VTI', 'VGK', 'VPL', 'VWO', 'PFF','IEI', 'BWX','AGG','BND','FXY','FXF','UDN','IAU','AOA','AOK','USO']
num = len(tickers)

data = pd.DataFrame()
high = pd.DataFrame()
low = pd.DataFrame()
close = pd.DataFrame()
data_id = tickers
#q
for i in tickers:
    tmp = web.DataReader(i, 'yahoo', '1/2/2014', '4/30/2019')   #1/3 is the first business day of 2005
    data[i] = tmp['Adj Close']
    high[i] = tmp['High']
    low[i] = tmp['Low']
    close[i] = tmp['Close']

```

```python
plt.figure(figsize=(12,7))
for i in tickers:
    plt.plot(data[i], label=i)
    
plt.legend(loc='upper right', fontsize=8)
plt.title("ETF Price")
plt.show()
```

```python
data.to_csv("data_adjclose.csv")
high.to_csv("data_high.csv")
low.to_csv("data_low.csv")
close.to_csv("data_close.csv")
```
