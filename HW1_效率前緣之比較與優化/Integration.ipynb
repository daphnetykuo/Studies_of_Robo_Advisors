```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
def round_list_num(firmlist):
    for i in range(len(firmlist)):
        firmlist[i] = round(firmlist[i],6)
    return firmlist

def to_float(firmlist):
    for i in range(len(firmlist)):
        firmlist[i] = float(firmlist[i])
    return firmlist

def get_ret_std(csvfile):
    with open(csvfile, encoding="utf-8") as file:
        reader = csv.reader(file)
        rows = [row for row in reader]
    file.close()
    ret, std = np.array(round_list_num(to_float(rows[0]))), np.array(round_list_num(to_float(rows[1])))
    return ret, std
```

```python
Betterment_ret, Betterment_std = get_ret_std('Betterment.csv')
CharlesSchwab_ret, CharlesSchwab_std = get_ret_std('CharlesSchwab.csv')
WiseBanyan_ret, WiseBanyan_std = get_ret_std('WiseBanyan.csv')
```

```python
plt.figure(figsize=(12,8))

plt.plot(Betterment_std, Betterment_ret,'-.', label='Betterment', linewidth=3)
plt.plot(CharlesSchwab_std, CharlesSchwab_ret,'--', label='Charles Schwab',linewidth=3)
plt.plot(WiseBanyan_std, WiseBanyan_ret,':', label='WiseBanyan',linewidth=3)

plt.grid(True)
plt.legend(loc='upper left', fontsize=12)
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.title('Efficient Frontier of Robo Advisors')
#plt.colorbar(label='Sharpe ratio')
#plt.scatter(WiseBanyan_std, WiseBanyan_ret,marker='.', label='WiseBanyan')
```
