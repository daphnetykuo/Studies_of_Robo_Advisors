```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import datetime as dt
```

```python
data = pd.DataFrame()
close = pd.DataFrame()
high = pd.DataFrame()
low = pd.DataFrame()
tickers = ['EWA', 'EWZ', 'EWC', 'FXI', 'EWQ', 'EWG', 'EWH', 'EWI', 'EWJ', 'EWY',
           'EWN', 'EWS', 'EZA', 'EWP', 'EWD', 'EWL', 'EWT', 'EWU', 'IVV']
q = len(tickers)
for i in tickers:
    tmp = web.DataReader(i, 'yahoo', '1/3/2005', '12/31/2019')   #1/3 is the first business day of 2005
    close[i] = tmp['Close']
    data[i] = tmp['Adj Close']
    high[i] = tmp['High']
    low[i] = tmp['Low']
```

```python
close.to_csv("data_close.csv")
data.to_csv("data_adjclose.csv")
high.to_csv("data_high.csv")
low.to_csv("data_low.csv")
```
