```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import datetime
```

```python
data_1 = pd.DataFrame()
tickers = ['EWA', 'EWZ', 'EWC', 'FXI', 'EWQ', 'EWG', 'EWH', 'EWI', 'EWJ', 'EWY',
           'EWN', 'EWS', 'EZA', 'EWP', 'EWD', 'EWL', 'EWT', 'EWU', 'IVV']
q = len(tickers)
records = 0
for i in tickers:
    tmp = web.DataReader(i, 'yahoo', '1/3/2005', '12/31/2012')   #1/3 is the first business day of 2005
    data_1[i] = tmp['Adj Close']
    if records == 0:
        records = len(tmp)
    else:
        records = min(records, len(tmp))

data_1.head()
```

```python
data_1.to_csv("data_firsthalf")
```

### View 1：以RSI預測ETF未來績效

```python
def cal_rsi(series):    #series: data_1["ticker"]
    up_days = series[series > 0]
    down_days = series[series < 0]
    up = up_days.mean() if len(up_days) > 0 else 0
    down = abs(down_days.mean()) if len(down_days) > 0 else 0
    return 100 * up / (up + down)
```
