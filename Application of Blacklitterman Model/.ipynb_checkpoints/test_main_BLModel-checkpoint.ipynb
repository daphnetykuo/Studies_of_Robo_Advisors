```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import datetime as dt
import statistics as stat
import csv
```

## 1.  資料匯入

```python
tickers = ['EWA', 'EWZ', 'EWC', 'FXI', 'EWQ', 'EWG', 'EWH', 'EWI', 'EWJ', 'EWY',
           'EWN', 'EWS', 'EZA', 'EWP', 'EWD', 'EWL', 'EWT', 'EWU', 'IVV']
```

```python
data = pd.read_csv("data_firsthalf.csv", index_col=0 )
data.index = pd.to_datetime(data.index)
data.head()
```

```python
change = data - data.shift(1)
change.head()
```

```python
returns = (data - data.shift(1)) / data.shift(1)
returns.head()
```

## 2. 兩種觀點下的long-short portfolio

### 技術指標1：14日RSI

```python
def cal_rsi(series):    #series: data["ticker"]
    up_days = series[series > 0]
    down_days = series[series < 0]
    up = up_days.mean() if len(up_days) > 0 else 0
    down = abs(down_days.mean()) if len(down_days) > 0 else 0
    return 100 * up / (up + down)
```

```python
rsi = pd.DataFrame()
for i in tickers:
    rsi[i] = change[i].rolling(14).agg(cal_rsi).dropna()   #14d-RSI
rsi
```

```python
monthly_rsi = rsi.asfreq('BM')
monthly_rsi.head()
```

### 定義函數：抓每個月RSI前四和後四檔的ETF 與 

```python
def get_8_etf_rsi(time):   #time: "y-m" ex."2005-1"   #output: 兩個list(long, short)
    temp_rsi_set = monthly_rsi[time]   #type: pd.dataframe
    #print(temp_rsi_set)
    #print(monthly_rsi.index[i])
    monthly_rsi_set = []   #每個月的etf排名及其rsi值
    for j in range(temp_rsi_set.shape[1]):   #df行數
        etf_rsi = [tickers[j], temp_rsi_set.iloc[0, j]]   #ix: label or index searching
        monthly_rsi_set.append(etf_rsi)
    monthly_rsi_set.sort(key=lambda x: x[1], reverse=True)
    #print(monthly_rsi_set)
    #取出每月前四檔和後四檔etf名稱
    long_4_etf = [k[0] for k in monthly_rsi_set[:4]]
    short_4_etf = [l[0] for l in monthly_rsi_set[-4:]]
    return long_4_etf, short_4_etf
```

### 使用函數跑每月的資料（2005-1~2012-12)

```python
weight_rsi = pd.DataFrame(columns=tickers,index=pd.date_range('2005/1/1','2012/12/31', freq='BM')).fillna(value=0)

for y in range(2005, 2013):
    for m in range(1, 13):
        time = str(y)+'-'+str(m)
        #print(time)
        tmp_etf = get_8_etf_rsi(time)
        for a in tmp_etf[0]:
            weight_rsi.loc[time, a] = 1
        for b in tmp_etf[1]:
            weight_rsi.loc[time, b] = -1

weight_rsi.head()
```

```python
q_rsi = []    #q1~q95
tmp_weight_rsi = weight_rsi.shift(1)   #把前個月的權重換成下個月index，使ret_下個月和weight_這個月配在一起
for y in range(2005, 2013):
    for m in range(1, 13):
        time = str(y)+'-'+str(m)   #這個月
        #print(time)
        if time != '2005-1':
            tmp_ret = returns[time].mean()    #series   #若沒有取mean,則為dataframe
            monthly_q_rsi = float(tmp_weight_rsi[time].dot(tmp_ret).to_numpy())
            q_rsi.append(monthly_q_rsi)
        else:
            pass
print(q_rsi)
print()
print('RSI mean:', stat.mean(q_rsi))
```

<b/>以RSI作為投資組合選擇指標，蒐集每月的報酬率（Rp_大 - Rp_小），則平均報酬率為-0.1658%。

### 技術指標2：移動平均線 MA

```python
ma_5d = pd.DataFrame()
for i in tickers:
    ma_5d[i] = data[i].rolling(5).mean().dropna()
ma_5d
```

```python
ma_20d = pd.DataFrame()
for i in tickers:
    ma_20d[i] = data[i].rolling(20).mean().dropna()
ma_20d
```

```python
monthly_ma = (ma_5d - ma_20d).asfreq('BM')
monthly_ma
```

若為正數，即表示當日短均線超過長均線，為看漲的股票；反之則看跌。

```python
def get_8_etf_ma(time):   #time: "y-m" ex."2005-1"   #output: 兩個list(long, short)
    temp_ma_set = monthly_ma[time]   #type: pd.dataframe
    monthly_ma_set = []   #每個月的etf排名及其ma值
    for j in range(temp_ma_set.shape[1]):   #df行數
        etf_ma = [tickers[j], temp_ma_set.iloc[0, j]]   #ix: label or index searching
        monthly_ma_set.append(etf_ma)
    monthly_ma_set.sort(key=lambda x: x[1], reverse=True)
    #print(monthly_ma_set)
    #取出每月前四檔和後四檔etf名稱
    long_4_etf = [k[0] for k in monthly_ma_set[:4]]
    short_4_etf = [l[0] for l in monthly_ma_set[-4:]]
    return long_4_etf, short_4_etf
```

```python
weight_ma = pd.DataFrame(columns=tickers,index=pd.date_range('2005/1/1','2012/12/31', freq='BM')).fillna(value=0)

for y in range(2005, 2013):
    for m in range(1, 13):
        time = str(y)+'-'+str(m)
        #print(time)
        tmp_etf = get_8_etf_ma(time)
        for a in tmp_etf[0]:
            weight_ma.loc[time, a] = 1
        for b in tmp_etf[1]:
            weight_ma.loc[time, b] = -1

weight_ma.head()
```

```python
q_ma = []    #q1~q95
tmp_weight_ma = weight_ma.shift(1)   #把前個月的權重換成下個月index，使ret_下個月和weight_這個月配在一起
for y in range(2005, 2013):
    for m in range(1, 13):
        time = str(y)+'-'+str(m)   #這個月
        #print(time)
        if time != '2005-1':
            tmp_ret = returns[time].mean()    #series   #若沒有取mean,則為dataframe
            monthly_q_ma = float(tmp_weight_ma[time].dot(tmp_ret).to_numpy())
            q_ma.append(monthly_q_ma)
        else:
            pass
print(q_ma)
print()
print('MA mean:', stat.mean(q_ma))
```

<b/>以MA作為投資組合選擇的指標，蒐集每月的報酬率（Rp_大 - Rp_小），則平均報酬率為-0.1320%。
